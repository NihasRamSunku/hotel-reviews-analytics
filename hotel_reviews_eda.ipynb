```python
monthly_score = df.groupby(
    df['Review_Date'].dt.to_period('M')
)['Reviewer_Score'].mean()

plt.figure()
monthly_score.plot()
plt.title("Average Reviewer Score Over Time")
plt.show()
```

```python
df['positive_length'] = df['Positive_Review'].astype(str).apply(len)
df['negative_length'] = df['Negative_Review'].astype(str).apply(len)

result = df.groupby(pd.cut(df['Reviewer_Score'], bins=[0,5,7,9,10]), observed=True)[['positive_length', 'negative_length']].mean()
result
```

```python
result.plot(kind='bar', figsize=(8,5))
plt.title("Average Review Length by Reviewer Score")
plt.ylabel("Average Number of Characters")
plt.xticks(rotation=0)
plt.show()
```

```python
df['Review_Month'] = df['Review_Date'].dt.to_period('M')

hotel_time = (
    df.groupby(['Hotel_Name', 'Review_Month'])['Reviewer_Score']
    .mean()
    .reset_index()
)
hotel_time.head()
```

```python
df['expectation_gap'] = df['Reviewer_Score'] - df['Average_Score']

df['expectation_gap'].describe()

hotel_gap = (
    df.groupby('Hotel_Name')['expectation_gap']
    .mean()
    .sort_values()
)

hotel_gap.head(10)   # Over-promising hotels
```

```python
hotel_gap.tail(10)   # Under-rated hotels
```

```python
hotel_variance = (
    df.groupby('Hotel_Name')['Reviewer_Score']
    .agg(['mean', 'std', 'count'])
)

polarized_hotels = hotel_variance[
    (hotel_variance['count'] >= 50) &
    (hotel_variance['std'] > 2)
].sort_values('std', ascending=False)

polarized_hotels.head(10)
```

```python
# Geographic Analysis: Location-based performance using latitude & longitude

import pandas as pd
import matplotlib.pyplot as plt

# -------------------------------
# 1. Aggregate performance by hotel location
# -------------------------------
geo_perf = (
    df.groupby(["Hotel_Name", "lat", "lng"])
      .agg(
          avg_reviewer_score=("Reviewer_Score", "mean"),
          avg_hotel_score=("Average_Score", "mean"),
          review_count=("Reviewer_Score", "count"),
          avg_expectation_gap=("expectation_gap", "mean")
      )
      .reset_index()
)

# -------------------------------
# 2. Scatter plot: Geographic distribution vs performance
# -------------------------------
plt.figure(figsize=(10, 6))
plt.scatter(
    geo_perf["lng"],
    geo_perf["lat"],
    s=geo_perf["review_count"] * 0.1,
)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Geographic Distribution of Hotel Performance")
plt.show()

# -------------------------------
# 3. Regional performance patterns using latitude bands
# -------------------------------
df["lat_band"] = pd.cut(
    df["lat"],
    bins=5
)

regional_perf = (
    df.groupby("lat_band")
      .agg(
          avg_reviewer_score=("Reviewer_Score", "mean"),
          avg_expectation_gap=("expectation_gap", "mean"),
          review_count=("Reviewer_Score", "count")
      )
      .reset_index()
)

print(regional_perf)

# -------------------------------
# 4. Identify high & low performing regions
# -------------------------------
high_performing_regions = regional_perf.sort_values(
    "avg_reviewer_score", ascending=False
)

low_performing_regions = regional_perf.sort_values(
    "avg_reviewer_score", ascending=True
)

print("High Performing Regions:")
print(high_performing_regions)

print("\nLow Performing Regions:")
print(low_performing_regions)
```
